```python
## 이상치제거
```

```python
df.boxplot(column='count')
```

```python
Q1 = df['amount'].quantile(q=0.25)
```

```python
Q3 = df['amount'].quantile(q=0.75)
```

```python
IQR = Q3-Q1
```

```python
IQR
```

```python
df1=df[(df['amount']<Q3+IQR*1.5)&(df['amount']>Q1-IQR*1.5)] #amount이상치 제거
df1
```

```python
df1_amount=df1['amount']
```

```python
df1_amount.hist()
```

```python
ndf = df.filter(['amount','count'])
```

```python
ndf=ndf[(ndf['count']<10)&(ndf['amount']<500000)]
ndf
```

```python
ndf.hist(bins=50,figsize=(20,15))
```

```python
# 앞쪽에 쏠려있는 금액 별수를 정규화 하기 위해 log를 씌우기
```

```python
import numpy as np
```

```python
ndf['logmount']=np.log(ndf['amount'])
```

```python
ndf.hist(bins=50,figsize=(20,15)) #logamount와 amount 확인하기
```

```python
# log를 취한건 한쪽으로 정렬됨. 데이터단위는 달라지지만 중심의 경향들이 파악될 수 있음
#이상치 제거하고 이상치제거된 부분도 쏠려있어서 log를 통해 정규화.
```

```python
sns.jointplot(x='amount',y='count',data=ndf) #이상치 제거하기 전에 상관계수는 더 작고, 유의하지도 않았음
              # 이상치 제거 후 상관관계가 유의하게 나타남 = 이상치가 크게 문제를 발생함
```

```python
sns.jointplot(x='amount',y='count',data=ndf, kind='kde')
```

```python
sns.jointplot(x='logmount',y='count',data=ndf, kind='kde')
```

```python
# logmount는 중심으로 쏠리고, 상관계수는 커지고, p값은 작아짐.
# 이러한 자료 처리가 되어야 분포에 대한 중심성이 갖춰져야 의미있는 데이터가 될수 있음.
```
