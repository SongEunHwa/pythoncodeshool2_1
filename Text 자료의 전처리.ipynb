## 02. 전처리를 위한 패키지 불러오기

```python
from nltk.tokenize import word_tokenize
```

```python
from gensim import corpora
```

### 03. 자료에서 초록만 뽑아오기

```python
title = data.ABSTRACT
```

```python
title
```

### 04. '_' 공백처리를 위한 함수

```python
doc_set=[]
```

```python
words=[]
```

```python
for doc in title :
    if type(doc) != float : #숫자가 아니면
        doc_set.append(doc.replace("_"," ")) # _를 공백으로 바꿔라
print("\n##### replace '_' to space #####\n")
print(doc_set)
```

### 05. 불용어, 어간추출

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stopWords = set(stopwords.words("english"))
```

```python
stopWords
```

```python
stemmer=PorterStemmer()
```

```python
for doc in doc_set :
    tokenizedWords=word_tokenize(doc.lower()) #소문자로 만들어 토큰화해서 넣음
    print(tokenizedWords)
    stoppedWords = [w for w in tokenizedWords if w not in stopWords] #stopWords에 있는 단어가 아닐때 w 불용어처리
    stemmedWors = [stemmer.stem(w) for w in stoppedWords] # 어간처리
    words.append(stemmedWors)
```

### 06. 전처리된 단어 및 corpus

```python
print(words)
```

#### 전처리 되지 않은 기호들 메모장에 정리

```python
dictionary = corpora.Dictionary(words) #토큰화 되어있는 걸 사전화 시켜라
```

```python
corpus = [dictionary.doc2bow(word) for word in words]
```

```python
print(corpus) #(몇번째단어,빈도)
```
